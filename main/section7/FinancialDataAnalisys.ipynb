## Price changes and financial returns

Price changes

```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.4f}'.format
plt.style.use('seaborn')
```

```python
close = pd.read_csv('close.csv',index_col="Date", parse_dates=["Date"])
close
```

```python
msft = close.loc[:,"MSFT"].dropna().to_frame().copy()
msft
```

```python
msft.rename(columns= {"MSFT":"Price"}, inplace = True)
msft
```

```python
msft["p_lag1"] = msft.shift(periods=1)
msft
```

### Absolute pricechanges

```python
#1 alternative:
msft["P_diff"] = msft.Price.sub(msft.p_lag1)
msft
```

```python
# altranative2
msft["P_diff2"]= msft.Price.diff(periods=1)
```

```python
msft
```

```python
msft.P_diff.equals(msft.P_diff2)
```

### REalative changes price

```python
#Alternative 1
msft.Price.div(msft.p_lag1) - 1
```

```python
# Alternative 2
msft["Returns"] = msft.Price.pct_change(periods = 1)
msft
```

```python
msft.drop(columns=["p_lag1", "P_diff", "P_diff2"], inplace=True)
msft
```

```python
msft.to_csv("msft.csv")
```

## Investment multiple and CAGR

```python
import pandas as pd
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.4f}'.format
plt.style.use('seaborn')
```

```python
msft = pd.read_csv("msft.csv", index_col="Date", parse_dates=["Date"])
msft
```

#### Investment multiple

```python
multiple =(msft.Price[-1]/ msft.Price[0])
multiple
```

#### Price incrace

```python
(multiple-1)*100
```

#### Compound Annual Growth Rate(CAGR)
https://pl.wikipedia.org/wiki/Skumulowany_roczny_wska%C5%BAnik_wzrostu

```python
start = msft.index[0]
start
```

```python
end = msft.index[-1]
end
```

```python
td = end - start
td
```

```python
td_years = td.days/365.25
td_years
```

```python
cagr = multiple**(1/td_years)-1 # short version
cagr
```

**CAGR canbe used to compare Investments with different investment horizons**

### Compound returns & geometric Meanreturns

```python
multiple = (1 + msft.Returns).prod() #3 alternative to calculate multiple
multiple
```

```python
n = msft.Price.count()
n
```

```python
geo_mean= multiple**(1/n) - 1 # Geometric mean return(daili)
geo_mean
```

```python
(1+geo_mean)**n    #4 alternate to calculate multiple
```

```python
mu = msft.Returns.mean()
mu
```

```python
(1+mu)**n
```
